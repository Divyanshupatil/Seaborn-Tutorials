# Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import / Load Data

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

# Distribution Plot

### It is mainly used for a single set of values (univariate data) and shows the data using a histogram. This means we look at just one column from the dataset.

```python
dis_plot = sns.distplot(crash_df['not_distracted'],kde = False)
dis_plot.set_title('Distribution Plot')
```

# Joint Plot

### A jointplot in Seaborn is used to show the relationship between two variables. It combines:A scatter plot (or other plot types) to show the relationship between the two variables.Histograms (or KDE plots) on the top and right to show the distribution of each variable individually.

```python
join_plot = sns.jointplot(x ='speeding',y='alcohol',data=crash_df,kind='reg') # this will create a joint plot

hex_plot = sns.jointplot(x = 'speeding',y='alcohol',data = crash_df,kind='hex') # this will create a hexagon distribution
```

# KDE Plot

### A KDE plot in Seaborn is a way to show how data is spread out. It gives a smooth curve instead of bars (like a histogram), which helps you see where most of the values are and how the data is shaped.

```python
kde_plot = sns.kdeplot(data = crash_df, x = 'alcohol')
kde_plot.set_title('KDE Plot')
```

# Pair Plot

### A pair plot in Seaborn is a great way to visualize the relationships between several columns in a dataset all at once. It creates a grid of plots where each variable is compared with every other variable using scatter plots, and the diagonal shows how each individual variable is distributed

```python
pair_plot = sns.pairplot(crash_df,hue='alcohol')
```

# Rug Plot

### A rug plot in Seaborn is a simple way to show the distribution of data using small vertical lines (like ticks) along the x-axis.Each line represents one data point, so you can easily see where the data is concentrated or spread out. It's often used along with other plots (like KDE or histograms) to give extra detail about individual values.

```python
rug_plot = sns.rugplot(data = crash_df, x = 'ins_premium')
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

# Bar Plot

### A bar plot in Seaborn is used to display the relationship between a categorical variable and a numerical variable. It shows the data as rectangular bars, where the length of each bar represents the value of the numerical variable for each category.

```python
bar_plot = sns.barplot(x='sex', y ='total_bill',data = tips_df)
```

# Count Plot

### A count plot in Seaborn is used to display the count (frequency) of observations in each category of a categorical variable. It’s similar to a bar plot, but instead of showing a numerical value, it shows the number of occurrences of each category.

```python
count_plot = sns.countplot(x='sex',data = tips_df, palette='Set2')
```

# Box Plot

### A box plot in Seaborn is used to show the distribution of a numerical variable and identify outliers, medians, and the spread of the data across categories. It gives a quick summary of the data using five key values: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

```python
box_plot = sns.boxplot(x='day',y = 'total_bill',data = tips_df,hue = 'sex')
```

# Violin Plot

### A violin plot in Seaborn is a combination of a box plot and a KDE plot (smoothed distribution). It not only shows the summary statistics (like median and quartiles) but also the shape of the data distribution.

```python
violin_plot = sns.violinplot(x='day',y = 'total_bill',data = tips_df,hue = 'sex')
```
